```python
import numpy as np

A = np.random.randint(10, size = (5, 5))
print("A")
print(A)
print("W")

W, V = np.linalg.eig(A)
print(W)
print("V")
print(V)
print("Vinv")

Vinv = np.linalg.inv(V)
print(Vinv)
print("B")

B = np.dot(np.dot(V, np.diag(W)), Vinv)
print(B)
print("S")

S = min(W)

print(S)
print("C")

w = W.copy()

for i in range(len(W)):
    w[i] = W[i] - S + 0.1


C = np.dot(np.dot(V, np.diag(w)), Vinv)
print(C)
print("NewEW")

eW, eV = np.linalg.eig(C)
print(eW)
print("Diff")

Diff = A - C
print(Diff)

EW, EV = np.linalg.eig()
```
